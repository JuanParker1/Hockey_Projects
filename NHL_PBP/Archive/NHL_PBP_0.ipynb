```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
item = []
broken = []
First = 'http://www.nhl.com/scores/htmlreports/'
Year = 20162017
Second = '/PL0'
Game = 20001
Third = '.HTM'
```

```python
while Game < 21280:
    url = First+str(Year)+Second+str(Game)+Third
    data = requests.get(url)
    soup = BeautifulSoup(data.text, 'html.parser')
    if data.status_code != 200:
        print(f'{Game}!!!')
        broken.append(Game)
    if data.status_code == 200:
        for tr in soup.find_all('tr',{'class':['evenColor','	oddColor','oddColor']}):
    #         for td in tr.find_all('td'):
    #             print(td)
            values = [td.text for td in tr.find_all('td',{'class':['+ bborder','+ bborder + rborder',' + bborder + rborder',' + bborder','goal + bborder','bold + bborder + rborder','bold + bborder','penalty + bborder','italicize + bold + bborder + rborder','italicize + bold + bborder']})]
            values.insert(0,'0'+str(Game))
            #print (values)
            item.append(values)
        item = [[x.replace('\n','') for x in l] for l in item]
        item = [[x.replace('\xa0',' ') for x in l] for l in item]
        item = [[x.replace(':','~!') for x in l] for l in item]
        print(Game)
    Game = Game+1
```

```python
df = pd.DataFrame(item)
brokendf = pd.DataFrame(broken)
```
